```python
#Encoding the categorical variables
from sklearn.preprocessing import LabelEncoder
def Label_Encoder(variable):
    le=LabelEncoder()
    return le.fit_transform(variable)
```

```python
bank['Gender']=Label_Encoder(bank['Gender'])
bank.head(5)
```

```python
bank['Attrition_Flag']=Label_Encoder(bank['Attrition_Flag'])
bank.head(5)
```

```python
bank['Education_Level']=Label_Encoder(bank['Education_Level'])
bank.head(5)
```

```python
bank['Marital_Status']=Label_Encoder(bank['Marital_Status'])
bank.head(5)
```

```python
bank['Income_Category']=Label_Encoder(bank['Income_Category'])
bank.head(5)
```

```python
bank['Card_Category']=Label_Encoder(bank['Card_Category'])
bank.head(5)
```

```python
x=bank.drop(['CLIENTNUM','Card_Category'],axis=1)
x.head(5)
```

```python
y=bank['Card_Category']
y.head(5)
```

```python
#Outliers
Q1 = x.quantile(0.25)
Q3 = x.quantile(0.75)
IQR = Q3 - Q1
UL = Q3 + 1.5*IQR
LL = Q1 -1.5*IQR
print(UL)
print('**********')
print(LL)
```

```python
x[((x<LL)|(x>UL))]=np.NaN
```

```python
((x>UL) | (x < LL)).sum().sum()
```

```python
#KNN Imputation method
from sklearn.impute import KNNImputer
x_impute=KNNImputer(n_neighbors=5)
output=pd.DataFrame(x_impute.fit_transform(x),columns=x.columns)
output.head()
```

```python
output.isnull().sum().sum()
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_scaled=pd.DataFrame(scale.fit_transform(output))
x_scaled
```

```python
#split train and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42,
                                                   stratify=y)
```

```python
## Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
pred_train_rf = rf.predict(x_train)
pred_test_rf = rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print("***************Confusion Matrix*********************")
print(confusion_matrix(y_train, pred_train_rf))
print("*******************************")
print(confusion_matrix(y_test, pred_test_rf))
```

```python
print("***************classification_report*********************")
print(classification_report(y_train, pred_train_rf))
print("*******************************")
print(classification_report(y_test, pred_test_rf))
```

```python
# Random forest with feature selection
from sklearn.feature_selection import RFE
rf1 = RandomForestClassifier()
feature_1 = RFE(estimator=rf1, n_features_to_select=17, step=1 )
feature_1 = feature_1.fit(x_train, y_train)
```

```python
pred_train_rf1 = feature_1.predict(x_train)
pred_test_rf1 = feature_1.predict(x_test)
```

```python
print("***************accuracy_score*********************")
print(accuracy_score(y_train, pred_train_rf1))
print("*******************************")
print(accuracy_score(y_test, pred_test_rf1))
```

```python
# cross validation method
from sklearn.model_selection import cross_val_score
rf_rfe_training = cross_val_score(feature_1, x_train, y_train, cv=10)
print(rf_rfe_training)
print()
print(rf_rfe_training.mean())
```
